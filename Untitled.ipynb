```python
import os
import matplotlib.pyplot as plt
import cv2
import numpy.random as npr
```

```python
nums = os.listdir('./data')
```

```python
cat = npr.choice(nums)
```

```python
fls = sorted(os.listdir('./data/%s' % cat))
```

```python
for f_name in fls:
    img = cv2.imread('./data/%s/%s' % (cat, f_name))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    print(f_name)
    plt.imshow(img)
    plt.show()
```
